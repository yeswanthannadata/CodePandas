```python
import pandas as pd
import numpy as np
raw_data = pd.read_excel('Instant Raw Report (11).xlsx')
raw_data['Timestamp'] = pd.to_datetime(raw_data['Timestamp'])
raw_data.set_index('Timestamp', inplace=True)
```

# Active Power consumption pattern

```python
import matplotlib.pyplot as plt
%matplotlib inline

raw_data[['Active Power kW']].plot(figsize=(20,10), linewidth=2, fontsize=20)
plt.xlabel('Timestamp', fontsize=20);
```
